```python
# Import Lib
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from  sklearn.feature_extraction.text import TfidfVectorizer
from  sklearn.svm import LinearSVC
from  sklearn.metrics import accuracy_score

# Data Perprocessing

# 1 step load the dataset to pandas data frame

row_mail_data= pd.read_csv('spam_ham_dataset.csv')
#print(row_mail_data)

# 2 step Replace a null value with a Null String.

mail_data=row_mail_data.where((pd.notnull(row_mail_data)),' ')
#print(mail_data.shape)
#print(mail_data.head()) sample data

# 3 step Label spam mail as ; and Non-sppam mail (ham) as 1;

mail_data.loc[mail_data['label']=='spam','label', ]= 0
mail_data.loc[mail_data['label']=='ham','label', ]= 1

# 4 step Separate the data as  Text And Label X --> Text & Y --> Label.

X=mail_data['text']
Y=mail_data['label']
#print(X)
#print("----------------")
#print(Y)

# Setp 5 Train Test Split Dataset

# Split the Data As Train data and test Data.

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.8,test_size=0.2,random_state=3)

# Step 6 Feature Extraction

# Transform the test data to feature vectors that can be used as input to the SVM model using TfidfVectorzer
# Convert the test to lower case letters

feature_extraction=TfidfVectorizer(max_df=1,stop_words='english',lowercase='True')
X_train_feature=feature_extraction.fit_transform(X_train)
X_test_feature=feature_extraction.transform(X_test)

# Step 7 Convert Y_Train and y_test values as a Interger Values

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')

# Step 8 Training the Model===> Support Vector Machine.

# Training the SVMm model with Training data

model=LinearSVC()
model.fit(X_train_feature,Y_train)

# Evaluation of the model.

# Step 9 Prediction on Training data

prediction_on_training_data=model.predict(X_train_feature)

accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)

print("Accurcy on Training data is : ",accuracy_on_training_data)

# Step 10 Prediction on Testining Data

prediction_on_testing_data=model.predict(X_test_feature)

accuracy_on_testing_data=accuracy_score(Y_test,prediction_on_testing_data)

print("\n Accurcy on Testing data is : ",accuracy_on_testing_data)

# ================ Prediction on New Mail ========================

input_mail=["Subject: photoshop , windows , office . cheap . main trending abasements darer prudently fortuitous undergone lighthearted charm orinoco taster railroad affluent pornographic cuvier irvin parkhouse blameworthy chlorophyll robed diagrammatic fogarty clears bayda inconveniencing managing represented smartness hashish academies shareholders unload badness danielson pure caffein spaniard chargeable levin"]

# Convert test to feature vector

input_mail_features=feature_extraction.transform(input_mail)

# Making Prediction

prediction=model.predict(input_mail_features)

print("\n",prediction)

if (prediction[0]==1):

    print("\n HAM Mail")
else:
    print("\n SAPM Mail")
```
